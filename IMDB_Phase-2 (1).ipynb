```python
#importing libraries
from pandasql import sqldf
import pandas as pd
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
#GETTING THE FIRST DATAFRAME(df1)

df1=pd.read_csv('data1.csv')
df1.head()
```

**1) Display all the details of movies created by directors Christopher and Matt Reeves.**

```python
christ="select * from df1 where Director1 like ('Christopher%')"
details=pysqldf(christ)
print(len(details))
details.tail()
```

```python
reeves="select * from df1 where Director1 like ('Matt Reeve%')"
details2=pysqldf(reeves)
print(len(details2))
details2.head()
```
